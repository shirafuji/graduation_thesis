```python
import matplotlib.pyplot as plt
# 散布図表示
size_predict = [1.3, 2.1, 2.9, 3.7, 5.2]
size_answer = [1, 2, 3, 4, 5]

plt.scatter(size_answer, size_predict)
plt.xlabel("size")
plt.ylabel("prediction")
plt.grid(True)
## y=xの直線をひく
x = []
for i in range(60):
    x.append(i*0.1)
y = []
for i in range(60):
    y.append(i*0.1)
plt.plot(x, y, color='black')
## 誤差-20%の直線を引く
y = []
for i in range(60):
    y.append(i*0.1*0.8)
plt.plot(x, y, color='red')
## 誤差+20%の直線を引く
y = []
for i in range(60):
    y.append(i*0.1*1.2)
plt.plot(x, y, color='red')
```
